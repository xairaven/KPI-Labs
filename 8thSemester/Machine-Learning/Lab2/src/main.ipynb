1. Відкрити та зчитати наданий файл з даними.

```python
import pandas as pd
df = pd.read_csv("data/Input_Lab-2_3_4_WQ-R.csv", sep=";")
```

тип у завантаженому наборі даних.


```python
num_records = df.shape[0]
num_fields = df.shape[1]
print(f"Кількість записів (рядків): {num_records}")
print(f"Кількість полів (стовпців): {num_fields}")
print(df.dtypes)
```

4. Додати до набору даних атрибут HighQuality, який дорівнює 1, якщо quality більше або дорівнює шести, та 0 в інших випадках.

```python
df["HighQuality"] = (df["quality"] >= 6).astype(int)
display(df.head(10))
```

5. Вивести перші 10 записів набору даних та видалити після цього атрибут quality з набору даних.

```python
df.drop(columns=["quality"], inplace=True)
display(df.head(10))
```

6. Перемішати набір даних та розділити його на навчальну (тренувальну)
та тестову вибірки, використовуючи функцію ShuffleSplit.

```python
from sklearn.model_selection import ShuffleSplit

# Перемішування та розбиття на тренувальну (80%) і тестову (20%) вибірки
splitter = ShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
train_idx, test_idx = next(splitter.split(df))

df_train = df.iloc[train_idx].reset_index(drop=True)
df_test = df.iloc[test_idx].reset_index(drop=True)

print(f"Розмір тренувальної вибірки: {df_train.shape[0]}")
print(f"Розмір тестової вибірки: {df_test.shape[0]}")
```

7. Використовуючи відповідні функції бібліотеки scikit-learn, збудувати класифікаційну модель логістичної регресії та навчити її на тренувальній вибірці, вважаючи, що цільова характеристика визначається стовпчиком HighQuality, а всі інші виступають в ролі вихідних аргументів.


```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Виділяємо цільову змінну та вхідні характеристики
X_train = df_train.drop(columns=["HighQuality"])
y_train = df_train["HighQuality"]
X_test = df_test.drop(columns=["HighQuality"])
y_test = df_test["HighQuality"]

# Нормалізація ознак
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Модель логістичної регресії
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train_scaled, y_train)
```

8. Обчислити класифікаційні метрики збудованої моделі для тренувальної
та тестової вибірки. Представити результати роботи моделі на тестовій
вибірці графічно.

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Передбачення
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)

# Вивід метрик
print("Класифікаційний звіт (тренувальна вибірка):")
print(classification_report(y_train, y_train_pred))

print("Класифікаційний звіт (тестова вибірка):")
print(classification_report(y_test, y_test_pred))

# Графічне представлення матриці помилок
fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred, ax=ax[0])
ax[0].set_title("Матриця помилок (Train)")
ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred, ax=ax[1])
ax[1].set_title("Матриця помилок (Test)")
plt.show()
```

9. З'ясувати вплив максимальної кількості ітерацій (від 5 до 5000) на результати класифікації. Результати представити графічно.


```python
import numpy as np

iters = np.linspace(5, 5000, 20, dtype=int)
train_scores = []
test_scores = []

for it in iters:
    model = LogisticRegression(max_iter=it, random_state=42)
    model.fit(X_train_scaled, y_train)

    train_scores.append(model.score(X_train_scaled, y_train))
    test_scores.append(model.score(X_test_scaled, y_test))

plt.figure(figsize=(8, 5))
plt.plot(iters, train_scores, label="Train Accuracy")
plt.plot(iters, test_scores, label="Test Accuracy")
plt.xlabel("max_iter")
plt.ylabel("Accuracy")
plt.legend()
plt.title("Вплив max_iter на точність моделі")
plt.show()
```

10. Проаналізувати ступінь впливу атрибутів на результат класифікації.
Збудувати класифікаційну модель логістичної регресії, залишивши від  3 до 5 найбільш важливих атрибутів та порівняти її результати із  моделлю з п. 7.

```python
import numpy as np

# Визначення важливості коефіцієнтів
feature_importance = np.abs(model.coef_).flatten()
sorted_idx = np.argsort(feature_importance)[-5:]  # Топ-5 ознак

top_features = X_train.columns[sorted_idx]
print("Найбільш значущі атрибути:", list(top_features))

# Навчання моделі тільки на вибраних атрибутах
X_train_selected = X_train_scaled[:, sorted_idx]
X_test_selected = X_test_scaled[:, sorted_idx]

model_selected = LogisticRegression(max_iter=1000, random_state=42)
model_selected.fit(X_train_selected, y_train)

# Порівняння якості моделі
print("Класифікаційний звіт для моделі з 5 атрибутами:")
print(classification_report(y_test, model_selected.predict(X_test_selected)))
```
